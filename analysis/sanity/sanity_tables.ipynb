For the table we use the following outrage definition:
Each station records hourly, if it fails, it does not record, meaning we can easily count the missing hours (data point), which is our availability. This sounds naive first, but we checked that when a station fails, it never records constantly zero.

```python
outages = station_outage_rate(dl)
```

```python
for id, s in station_names.items():
    data = dl.get_bicycle(s, interval, "1d").to_pandas()
    avg_drivers = int(data["channels_all"].mean())
    avail = 100 * (1 - outages.filter(outages["station"] == s)["outage_rate"][0])
    
    print(f"{id} & {s} & {avg_drivers} & {avail:.2f} \\\\")
```
